```python
# std
import os
import sys
import inspect
import time
import pathlib
import glob
from math import sqrt
from math import log2
# packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
%matplotlib inline

# packages
from matplotlib.colors import ListedColormap
```

```python
def print_styling(figsize=(10,8)):
    plt.rc('figure', figsize=figsize) 
    plt.style.use("seaborn")
    sns.set_palette("dark")
    SMALL_SIZE = 15
    MEDIUM_SIZE = 18
    BIGGER_SIZE = 26

    plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
    plt.rc('lines', linewidth=2)

    plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
    plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
    plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
    plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
    
    plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
    plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
def plot_CV_with_Std(df, x, y, lable, xlable = "time", ylable = "time", titel = "R2-Score", SaveName = False, ax=None):
    if ax==None:
        ax = plt.gca()
    plt.sca(ax)

    plt.plot(df[x], df[y], '-', label = lable)

    plt.grid(True)
    plt.legend()
    plt.xlabel(xlable)
    plt.ylabel(ylable)
    plt.suptitle(titel)
    #plt.ylabel(y)
    if SaveName:
        plt.savefig(SaveName)
```

```python
df_results_MNIST = pd.read_csv("out/Results_MNIST.csv")
df_results_FASION = pd.read_csv("out/Results_FASION.csv")
print(df_results_MNIST.head())
```

```python
mpl.style.use('seaborn')
plt.figure(figsize=(10,8))
print_styling()
plot_CV_with_Std(df_results_MNIST,"epoch","accuracy","MNIST",xlable = "epoch",ylable = "accuracy", SaveName = "out/MNIST_accuracy", titel = "Accuracy")
plot_CV_with_Std(df_results_FASION,"epoch","accuracy","FASHION",xlable = "epoch",ylable = "accuracy", SaveName = "out/MNIST_accuracy", titel = "Accuracy")
```
